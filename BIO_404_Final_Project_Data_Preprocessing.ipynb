```python
import pandas as pd
```

```python
df = pd.read_csv("oasis_cross-sectional.csv")
df
```

```python
df.info()
```

```python
# Drop columns that aren't needed
df = df.drop("ID", axis=1)
df = df.drop("Delay", axis=1)  # As seen above, this column mostly has null data
df = df.dropna()  # Drop any records left with null data 

# Normalize Alzheimer's classifications to have an equal difference between them
df["CDR"] = df["CDR"].replace([0, 0.5, 1, 2],[0, 1, 2, 3])  

# Rearranging columns to have classification as first column
df = df[['CDR', 'M/F', 'Hand', 'Age', 'Educ', 'SES', 'MMSE', 'eTIV', 'nWBV','ASF']] 
df
```

```python
df.info()
```

```python
df.to_csv("processed_oasis_cross-sectional.csv", index=False)  # Save as CSV
```
